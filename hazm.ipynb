```python
import pandas as pd
# we separate data with  " to get access to titles and contents
rawDf = pd.read_csv('ferdowsi-data.txt',names = ['col1', 'col2', 'col3', 'col4'],sep = '"', nrows = 32)# couldnt run local more lines and time was over :))
rawDf
```

<div dir='rtl'>اکنون تمام کانتنت ها و تایتل ها را در یک استرینگ ذخیره میکنیم</div>

```python
articleContent = []
index = list(rawDf.index)
for i in range(len(index)):
    try:
        if( ("content" in rawDf.loc[i, 'col2']) | ("title" in rawDf.loc[i, 'col2']) ):
            articleContent.append(rawDf.loc[i, 'col4'])
    except:
        pass
```

<div dir='rtl'>حال تمام داده های مورد نیاز را داریم ولی مشکل بزرگی که با آن روبه رو هستیم این است که این داده ها قابل خواندن نیستند. برای حل این مشکل کافیست جفت بک اسلش ها را تبدیل به یک بک اسلش بکنیم. به همین راحتی :))</div>

```python
import ast
RarticleContent = articleContent
for i in range(len(RarticleContent)):
    try:
        RarticleContent[i] = ast.literal_eval(RarticleContent[i].join("'" * 2))
    except:
        pass
```

```python
RarticleContent = ''.join(RarticleContent)
RarticleContent
```

<div dir="rtl">
<h2>
نورمالایز کردن داده ها </h2>
</div>
<hr/>

<div dir='rtl'>کتابخوانه هضم را ایمپورت میکنیم و اقدام به نومالایز کردن داده میکنیم.</div>

```python
from __future__ import unicode_literals
from hazm import *
normalizer = Normalizer()
NarticleContent = normalizer.normalize(RarticleContent)
```

<div dir="rtl">
<h2>
توکنایز کردن داده ها </h2>
</div>
<hr/>

<div dir='rtl'>با استفاده از همان کتابخوانه هضم اقدام به توکنایز کردن داده میکنیم.</div>

```python
articleTokenize = word_tokenize(NarticleContent)
```

<div dir="rtl">
<h2>
ده کلمه پر تکرار مقاله ها </h2>
</div>
<hr/>

<div dir='rtl'>حال که داده ها را توکنایز کردیم تنها کافیست اقدام به مشارش داده های یونیک کنیم.</div>

```python
import numpy as np
numpyArt = np.array(articleTokenize)
unique, counts = np.unique(numpyArt, return_counts=True)
numpyArt = dict(zip(unique, counts))

```


<div dir='rtl'>برخی از کاراکتر ها هستند که اضافی هستن اما تکنایز شده بین داده های ما وجود دارند، با استفاده از یک حلقه فور ساده از شر آنها خلاص میشویم.</div>

```python
errors = ['.', ':', ':', '(', ')', '«', ',', '»', '،', '؛', '؟']
for i in errors:
    try:
        del numpyArt[i]
    except:
        pass
```


<div dir='rtl'>حال تقریبا همه چیز آماده است، تنها کاری که باید بکنیم اینه که ولیو ها را سورت کنیم تا بیشترین کلمه ها را پیدا کنیم.</div>

```python
# sorting dictionary by values
values = []
for x, y in numpyArt.items():
    values.append(y)
values.sort(reverse=True)
values = values[0:10]
# Let's print the resutl
for x, y in numpyArt.items():
    if (y in values):
        print(x)
```

<div dir="rtl">
<h2>
ده بایگرام پر تکرار مقاله ها </h2>
</div>
<hr/>

<div dir='rtl'>الگوریتمی که پیش میگیریم اینگونه هست که در داده ای که در لیست توکنایز شده را داری، حروفش را دوتا دوتا جدا میکنیم و در یک لیست جدید ذخیره میکنیم.</div>

```python
bigrams = []
for i in articleTokenize:
    for j in range(len(articleTokenize)):
        try:
            if(len(i[j:j+2]) == 2):
                bigrams.append(i[j:j+2])
        except:
            pass
```


<div dir='rtl'>حالی که در یک لیست همه بایگرام هارا داریم، تعداد آن هارا میشماریم و در یک دیکشنری ذخیره میکنیم.</div>

```python
import numpy as np
numpyBi = np.array(bigrams)
unique, counts = np.unique(numpyBi, return_counts=True)
numpyBi = dict(zip(unique, counts))
```


<div dir='rtl'>حال تنها کافیت آنها را با توجه به تعداد تکرارشون سورت کنیم و نتیجه را برگردانیم.</div>

```python
# sorting the dictionary
values = []
for x, y in numpyBi.items():
    values.append(y)
values.sort(reverse=True)
values = values[0:10]
# printing the result
for x, y in numpyBi.items():
    if (y in values):
        print(x)
```
